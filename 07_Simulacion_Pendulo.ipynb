```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definicin de las constantes del pndulo
g = 9.81  # Aceleracin debida a la gravedad en m/s^2
L = 1.0   # Longitud del pndulo en metros

# Ecuacin diferencial del pndulo
def pendulum(y, t, L, g):
    theta, omega = y
    dtheta_dt = omega
    domega_dt = -(g / L) * np.sin(theta)
    return [dtheta_dt, domega_dt]

# Condiciones iniciales (ngulo inicial de 45 grados, velocidad angular inicial cero)
y0 = [np.pi / 4, 0.0]

# Creacin de un arreglo de tiempo desde 0 hasta 10 segundos
t = np.linspace(0, 10, 1000)

# Resolver la ecuacin diferencial
solution = odeint(pendulum, y0, t, args=(L, g))

# Extraccin de los ngulos y las velocidades angulares de la solucin
theta = solution[:, 0]
omega = solution[:, 1]

# Creacin de grficos para ngulo y velocidad angular
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(t, theta)
plt.title('ngulo vs. Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('ngulo (rad)')

plt.subplot(1, 2, 2)
plt.plot(t, omega)
plt.title('Velocidad Angular vs. Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad angular (rad/s)')

plt.tight_layout()
plt.show()
```
